# Explore fat-tailed and non-fat-tailed processes :)

## Run these functions with `shift+enter` to run them

```python

import numpy as np
import matplotlib.pyplot as plt

def make_stick_plot( numbers ):
    fig = plt.figure(figsize=(len(numbers)/50., 5))
    plt.gca().vlines(range(len(numbers)), [0], numbers)
    fig.patch.set_visible(False)
    plt.gca().axis('off')

def make_stick_plot__bell_curve( n_numbers, standard_deviation=1. ):
    data = np.random.normal(loc=0, scale=standard_deviation, size=int(n_numbers) )
    make_stick_plot( data )
    
def make_stick_plot__fat_tail_curve( n_numbers, tail_fatness=2. ):
    data = np.random.zipf(tail_fatness, size=int(n_numbers) )
    make_stick_plot( data )

```

```python
n_vals, std_vals, avg_vals = [], [], []
for n in np.logspace(1, 6):
    data = np.random.zipf(1.5, int(n) )
    n_vals += [n]
    std_vals += [np.std( data )]
    avg_vals += [np.mean( data )]
plt.loglog( n_vals, avg_vals, 'b' )
plt.loglog( n_vals, std_vals, 'r' )
plt.xlabel(r'Number of samples $n$')
plt.title(r'average (blue), standard deviation (red)' )
```
