# Explore fat-tailed and non-fat-tailed processes :)

## Run these functions with `shift+enter` to run them

```python

import numpy as np
import matplotlib.pyplot as plt

def make_stick_plot( numbers ):
    fig = plt.figure(figsize=(len(numbers)/50., 5))
    plt.gca().vlines(range(len(numbers)), [0], numbers)
    fig.patch.set_visible(False)
    plt.gca().axis('off')

def make_stick_plot__bell_curve( n_numbers, standard_deviation=1. ):
    data = np.random.normal(loc=5, scale=standard_deviation, size=int(n_numbers) )
    make_stick_plot( data )
    
def make_stick_plot__fat_tail_curve( n_numbers, tail_fatness=2. ):
    data = np.random.zipf(tail_fatness, size=int(n_numbers) )
    make_stick_plot( data )

```


## These are random numbers from the Bell curve that we disussed
### Play around with the number of numbers (first change, then `shift+enter`)

```python
n_numbers = 100
make_stick_plot__bell_curve( n_numbers )
```

### If you don't want to figure out how to change the notebook, here are some typical snapshot

```python
make_stick_plot__bell_curve( 10 )
make_stick_plot__bell_curve( 100 )
make_stick_plot__bell_curve( 1000 )
```

## But now let's make graphs like that for a fat tailed distribution

```python
make_stick_plot__fat_tail_curve( 20, 2 )
```

### Looks totally fine, right? But let's sample more often...

**For this one, the largest stick is already 2x as big as the next one**

```python

make_stick_plot__fat_tail_curve( 100, 2 )
```

**But sampling more often makes the difference even bigger!**

```python
make_stick_plot__fat_tail_curve( 500, 2 )
```

**And even bigger...**

```python
make_stick_plot__fat_tail_curve( 2000, 1.2 )
```

## But this is a visualization, we can't yet see the average directly
## So let's plot the average

```python
def plot_average__bell_curve(n_samples_min_log=1, n_samples_max_log=5):
    n_vals, std_vals, avg_vals = [], [], []
    for n in np.logspace(n_samples_min_log, n_samples_max_log):
        data = np.random.normal(loc=1, scale=2, size=int(n) )
        n_vals += [n]
        std_vals += [np.std( data )]
        avg_vals += [np.mean( data )]
    plt.loglog( n_vals, avg_vals, 'b' )
    plt.plot( n_vals, std_vals, 'r' )
    plt.xlabel(r'Number of samples $n$')
    plt.title('Average from Bell curve \n average (blue), standard deviation (red)' )
    
def plot_average__fat_tail(n_samples_min_log=1, n_samples_max_log=6):
    n_vals, std_vals, avg_vals = [], [], []
    for n in np.logspace(n_samples_min_log, n_samples_max_log):
        data = np.random.zipf(1.5, int(n) )
        n_vals += [n]
        std_vals += [np.std( data )]
        avg_vals += [np.mean( data )]
    plt.loglog( n_vals, avg_vals, 'b' )
    plt.plot( n_vals, std_vals, 'r' )
    plt.xlabel(r'Number of samples $n$')
    plt.title('Average from Bell curve \n average (blue), standard deviation (red)' )
```

```python
plot_average__bell_curve(1,6)
```

```python
plot_average__fat_tail(1,7)
```
